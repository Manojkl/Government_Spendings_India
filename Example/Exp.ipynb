```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Wikipedia URL
url = "https://en.wikipedia.org/wiki/List_of_districts_in_India"

# Send a request to fetch the page content
response = requests.get(url)
response.raise_for_status()  # Ensure the request was successful

# Parse the HTML content
soup = BeautifulSoup(response.text, "html.parser")

# Find all relevant tags: <h3> for state/UT names and <table> for district data
elements = soup.find_all(
    ["h3", "table"]
)  # Get all headings and tables in sequential order

state_name = None  # To store the current state name

# Iterate through elements in sequence
for element in elements:
    if element.name == "h3":  # If the element is a state/UT heading
        state_name = element.get_text(strip=True).replace("[edit]", "").strip()

    elif (
        element.name == "table" and state_name
    ):  # If it's a table and a state name was found before
        try:
            # Convert the HTML table to a DataFrame
            df = pd.read_html(str(element))[0]

            # Create a valid filename
            filename = (
                f"{state_name.replace(' ', '_').replace('(', '').replace(')', '')}.csv"
            )

            # Save the table as a CSV file
            df.to_csv(filename, index=False, encoding="utf-8")
            print(f"Saved: {filename}")

        except Exception as e:
            print(f"Error parsing table for {state_name}: {e}")

print("All district tables have been successfully saved.")
```

```python
import os
import pandas as pd

# List of folder names (state names formatted)
updated_names_states = [name.replace(" ", "_") for name in states_of_india]

# Paths
base_directory = "/Users/manojkl/Documents/Government_Spendings_India/States/"  # Destination for folders
csv_directory = "/Users/manojkl/Documents/Government_Spendings_India/Example/"  # Folder containing CSV files

# Create state folders
for folder in updated_names_states:
    folder_path = os.path.join(base_directory, folder)
    os.makedirs(folder_path, exist_ok=True)  # Creates folder if it doesn't exist

print("State folders created successfully!")

# Process CSV files
for csv_file in os.listdir(csv_directory):
    if csv_file.endswith(".csv"):
        # Extract state name by removing _XX (where XX is the state code) from the filename
        state_name = "_".join(csv_file.split("_")[:-1])  # Removes last underscore part

        # Ensure the filename matches a created folder
        if state_name in updated_names_states:
            csv_path = os.path.join(csv_directory, csv_file)
            folder_path = os.path.join(base_directory, state_name)

            # Read CSV file
            df = pd.read_csv(csv_path)

            # Ensure the column "District[9]" exists
            district_column = next(
                (col for col in df.columns if "District" in col), None
            )
            if district_column:
                districts = (
                    df[district_column].dropna().unique()
                )  # Get unique district names

                # Create district folders inside the state folder
                for district in districts:
                    district_folder = os.path.join(
                        folder_path, district.replace(" ", "_")
                    )
                    os.makedirs(district_folder, exist_ok=True)

                print(f"District folders created under {state_name}")

print("All district folders created successfully!")
```

```python
import os
import pandas as pd

# Define paths
csv_directory = "/Users/manojkl/Documents/Government_Spendings_India/Example/"
ut_directory = "/Users/manojkl/Documents/Government_Spendings_India/Union_Territories/"

# Iterate through all CSV files
for csv_file in os.listdir(csv_directory):
    if csv_file.endswith(".csv"):
        # Extract state/UT name by removing "_XX" (last underscore segment)
        ut_name = "_".join(
            csv_file.split("_")[:-1]
        )  # Gets everything before the last "_XX"

        # Path to matching folder
        ut_folder_path = os.path.join(ut_directory, ut_name)

        # Check if the folder exists
        if os.path.isdir(ut_folder_path):
            # Read CSV file
            csv_path = os.path.join(csv_directory, csv_file)
            df = pd.read_csv(csv_path)

            # Convert CSV to Markdown format
            md_content = df.to_markdown(index=False)

            # Markdown file path
            md_file_path = os.path.join(ut_folder_path, f"{ut_name}.md")

            # Save to .md file
            with open(md_file_path, "w", encoding="utf-8") as md_file:
                md_file.write(f"# {ut_name} Union Territory\n\n")
                md_file.write(md_content)

            print(f"Markdown file created: {md_file_path}")
        else:
            print(f"Folder not found for: {ut_name}")

print("CSV to Markdown conversion completed successfully!")
```

```python
import os
import zipfile

# Define the folder containing CSV files
folder_path = "/Users/manojkl/Documents/Government_Spendings_India/Example/"

# Define the output ZIP file name
zip_file_path = os.path.join(folder_path, "archived_csvs.zip")

# Create a ZIP file
with zipfile.ZipFile(zip_file_path, "w", zipfile.ZIP_DEFLATED) as zipf:
    for file_name in os.listdir(folder_path):
        if file_name.endswith(".csv"):
            file_path = os.path.join(folder_path, file_name)
            zipf.write(file_path, arcname=file_name)  # Add file to zip
            os.remove(file_path)  # Delete the original CSV file

print(f"All CSV files are zipped into {zip_file_path} and deleted successfully.")
```
